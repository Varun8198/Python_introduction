# Question - 3:

### a) Write a class named Marsupial that can be used:
### b) Now write a class named Kangaroo as a subclass of Marsupial that inherits all the attributes of Marsupial.

```python
class Marsupial:
    def __init__(self):
        self.in_pouch = []     #Empty list where everything will be placed.
    def put_in_pouch(self,toy):
        self.in_pouch.append(toy)    #Everytime something is written, it will be appended to the list
    def pouch_contents(self):
        return self.in_pouch
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('Fire truck')
m.put_in_pouch('kitten')
m.pouch_contents()
```

```python
class Kangaroo(Marsupial):
    
    def __init__(self,x,y):
        Marsupial.__init__(self)   #Inherits all the atributes from Marsupial  
        self.x = x
        self.y = y
        
    def jump(self,dx,dy):
        self.x += dx          
        self.y += dy
        
    def __str__(self):
        return f'I am a Kangaroo located at coordinates {self.x,self.y}'
    
k = Kangaroo(0,0)
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
k.jump(1,5)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
```

# Question - 7:
Implement recursive function webdir() that takes as input: a URL (as a string) and
non-negative integers depth and indent. Your function should visit every web
page reachable from the starting URL web page in depth clicks or less, and print
each web page's URL. As shown below, indentation, specified by indent, should
be used to indicate the depth of a URL.

```python
from html.parser import HTMLParser
import urllib.request

class WebParser(HTMLParser):   
    def __init__(self, depth, indent):
        HTMLParser.__init__(self)
        self.depth = depth
        self.indent = indent
        self.currentDepth = 0

    def handle_starttag(self, tag, attrs): # Check if the tag is an anchor tag and has an href attribute
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    url = attr[1]    # Print the URL with appropriate indentation
                    print('  ' * self.currentDepth + url)
                    
    def handle_data(self, data): # Increment the current depth
        self.currentDepth += 1
        
    def handle_endtag(self, tag):   # Decrement the current depth by 1
        self.currentDepth -= 1
        
def webdir(url, depth, indent):  # Create an instance of the WebDirParser class
    hp = WebParser(depth, indent)   # Open the URL and read its contents
    response = urllib.request.urlopen(url)
    html = response.read()     # Feed the HTML content to the parser
    hp.feed(html.decode('utf-8'))

webdir('https://www.w3schools.com/python/python_inheritance.asp', 2, 0)


# The website you provided is throwing an error of HTTP error 404 (not found) when opened in browser also. 
# Therefore I used different websites to show that it is working.
```

Write SQL queries on Database Table:

```python
import sqlite3
conn = sqlite3.connect('database.db')
cursor = conn.cursor()
```

```python
# a) All the temperature data
cursor.execute("SELECT Temperature FROM table")
temperatureData = cursor.fetchall()
print(temperatureData)

# b) All the cities, but without repetition.
cursor.execute("SELECT DISTINCT City FROM table")
cities = cursor.fetchall()
print(cities)

# c) All the records for India
cursor.execute("SELECT * FROM table WHERE Country = 'India'")
records = cursor.fetchall()
print(records)

# d) All the Fall records
cursor.execute("SELECT * FROM table WHERE Season = 'Fall'")
fallRecords = cursor.fetchall()
print(fallRecords)

# e) The city, country, and season for which the average rainfall is between 200 and 400 millimeters
cursor.execute("SELECT City, Country, Season FROM table GROUP BY City, Country, Season HAVING AVG(rainfall) BETWEEN 200 AND 400")
averageRainfall = cursor.fetchall()
print(averageRainfall)

# f) The city and country for which the average Fall temperature is above 20 degrees, in increasing temperature order
cursor.execute("SELECT City, Country FROM table WHERE Season = 'Fall' GROUP BY City, Country HAVING AVG(temperature) > 20 ORDER BY AVG(temperature) ASC")
average_fall_temperature_records = cursor.fetchall()
print(average_fall_temperature_records)

# g) The total annual rainfall for Cairo
cursor.execute("SELECT SUM(Rainfall) AS total_Rainfall FROM table WHERE City = 'Cairo'")
cairoRainfall = cursor.fetchone()[0]
print(cairoRainfall)

# h) The total rainfall for each season
cursor.execute("SELECT Season, SUM(Rainfall) AS total_Rainfall FROM table GROUP BY Season")
seasonalRainfall = cursor.fetchall()
print(seasonalRainfall)

conn.close()
```
